## Ejercicio 2 

### 5.5 Importar datos desde una API-GET

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
from IPython.display import display

# 1. URL de la API (sin clave)
url = "https://akabab.github.io/superhero-api/api/all.json"

# 2. Petición a la API
response = requests.get(url)

# 3. Verificamos que la respuesta sea exitosa
if response.status_code == 200:
    data = response.json()

    # 4. Convertimos a DataFrame
    df = pd.json_normalize(data)

    # 5. Filtramos solo personajes de Marvel
    df_marvel = df[df['biography.publisher'] == 'Marvel Comics']

    # 6. Mostramos las primeras filas con columnas clave
    display(df_marvel[['name', 'biography.fullName', 'appearance.gender', 'powerstats.intelligence', 'powerstats.strength', 'powerstats.power']].head(10))

else:
    print(f"Error {response.status_code}: {response.text}")
```
